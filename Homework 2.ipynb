__2.4 (mammals data) Refer to Example 2.7. Create a new variable r equal to the ratio of brain size over body size. Using the full mammals data set, order the mammals data by the raio r. Which mammals have the largest ratio of brain size to body size? Which mammals have the smallest ratio?__

```r
library(MASS)
r = round(mammals$brain/mammals$body, digit = 5)
newMammals = cbind(mammals, r)
cat("Mammals have the largest ratio of brain size to body size is: ")
newMammals[which(newMammals$r == max(newMammals$r)),0]

cat("Mammals have the smallest ratio of brain size to body size is: ")
newMammals[which(newMammals$r == min(newMammals$r)),0]

```

__2.7 (Central Limit Theorem with simulated data). Refer to Example
2.6, where we computed sample means for each row of the randu data frame.
Repeat the analysis, but instead of randu, create a matrix of random numbers
using runif.__|

```r
r = matrix(runif(400*3, 0,1),nrow = 400, ncol = 3)
colnames(r) = c("x","y","z")
mean(r)
var(r)
diag(var(r))
cor(r)
means = apply(r, MARGIN = 1, FUN = mean)
hist(means)
```

__2.13 (mammals data on original scale).__ Refer to the mammals data in Example 2.7. Construct a scatterplot like Figure 2.19 on the original scale (Figure 2.19 is on the log-log scale.) Label the points and distances for cat,
cow, and human. In this example, which plot is easier to interpret?

```r
plot(brain ~ body, mammals, xlab = "Body", ylab = "Brain")
y = mammals[c("Cat", "Cow", "Human"),]
polygon(y)
text(y, rownames(y))
dist(y)
     
```
