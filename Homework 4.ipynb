__4.2 (Speed and stopping distance (continued)).__ <br> Suppose that one wishes to compare linear and quadratic fits to the (speed,dist) observations. One can construct these two fits in R using the code <br>

fit.quadratic = lm(dist ~ speed + I(speed^2), data=cars)<br>

__a.__ Construct a scatterplot of speed and stopping distance. <br>
__b.__ Using the abline function with argument fit.linear, overlay the best line fit using line type “dotted” and using a line width of 2<br>
__c.__ Using the lines function, overlay the quadratic fit using line type “longdash” and a line width of 2.<br>
__d.__ Use a legend to show the line types of the linear and quadratic fits.<br>
__e.__ Redo parts (a) - (d) using two contrasting colors (say red and blue) for
the two different fits.

````r
fit.linear = lm(dist ~ speed, data=cars)
fit.linear
````

````r
plot(dist ~ speed, data  = cars, xlab = "Speed(mpg)", ylab = "Stopping Distance (ft)", 
                             main = "Speed vs. Distance", col = "Red", pch = 17)
abline(fit.linear, lty = "dotted", lwd = 2)
````

````r
fit.quadratic = lm(dist ~ speed + I(speed^2), data=cars)
fit.quadratic
xval = speed
hat2 = predict(fit.quadratic, newdata=list(speed=xval))
````

````r
plot(dist ~ speed, data  = cars, xlab = "Speed(mpg)", ylab = "Stopping Distance (ft)", 
                             main = "Speed vs. Distance", col = "Red", pch = 17)
abline(fit.linear, lty = "dotted", lwd = 2, col = "red")
lines(xval, hat2, col="blue", lty= "longdash", lwd=2)

legend("topleft", legend = c("Linear", "Quadratic"), lty = c("dotted", "longdash", col = c("red", "blue")))
````

__4.7 ( lattice graphics).__ The dataset faithful contains the duration of the eruption (in minutes) eruptions and the waiting time until the next eruption waiting (in minutes) for the Old Faithful geyser. One is interested in exploring the relationship between the two variables. <br>
__a.__ Create a factor variable length that is “short” if the eruption is smaller than 3.2 minutes, and “long” otherwise. <br> 
faithful\$length = ifelse(faithful\$eruptions < 3.2, "short", "long")<br>
__b.__ Using the bwplot function in the lattice package, construct parallel boxplots of the waiting times for the “short” and “long” eruptions.<br>
__c.__ Using the densityplot function, construct overlapping density plots of the waiting times of the “short” and “long” eruptions <br>

````r
library(lattice)
faithful$length = ifelse(faithful$eruptions < 3.2, "short", "long")
bwplot(faithful$waiting ~ faithful$length)
````

````r
densityplot(~ faithful$waiting, groups= faithful$length , data=faithful, auto.key=list(space="top"))
````
