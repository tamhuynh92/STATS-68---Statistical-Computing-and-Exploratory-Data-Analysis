__5.2 (Relationship between the percentages of small classes and large classes).__ The variables Pct.20 and Pct.50 in the college dataset contain respectively the percentage of “small classes” (defined as 20 or fewer students) and the percentage of “large classes” (defined as 50 or more students) in the National Universities. <br>
__a.__ Use the plot function to construct a scatterplot of Pct.20 (horizontal) against Pct.50 (vertical).<br>
__b.__ Use the line function to find a resistant line to these data. Add this resistant line to the scatterplot constructed in part a.<br>
__c.__ If 60% of the classes at a particular college have 20 or fewer students, use the fitted line to predict the percentage of classes that have 50 or more students.<br>
__d.__ Construct a graph of the residuals (vertical) against Pct.20 (horizontal) and add a horizontal line at zero (using the abline function).<br>
__e.__ Is there a distinctive pattern to the residuals? (Compare the sizes of the residuals for small Pct.20 and the sizes of the residuals for large Pct.50.)<br>
__f.__ Use the identify function to identify the schools that have residuals that exceed 10 in absolute value. Interpret these large residuals in the context of the problem.<br>

````r
dat = read.table("college.txt", header=TRUE, sep="\t")
college = subset(dat, complete.cases(dat))
plot(x = college$Pct.20, y = college$Pct.50, xlab = "Small Classes", ylab = "Large Classes")
fit = line(college$Pct.20, college$Pct.50)
abline(coef(fit), col = "red")
box()
````

````r
percentage.Pct20 = 0.60
coeffs = coefficients(fit)
cat("If 60% of the classes at a particular college have 20 or fewer students,"  
         ,"the percentage of classes that have 50 or more students is: ", percentage.Pct20 * coeffs[1] + coeffs[2])
````

````r
plot(x = college$Pct.20, y = fit$residual, xlab = "Small Classes", ylab = "Residuals")
abline(h=0, col = "red")
````

````r
plot(x = college$Pct.50, y = fit$residual, xlab = "Large Classes", ylab = "Residuals")
abline(h=0, col = "red")
````

````r
plot(x = college$Pct.20, y = fit$residual, xlab = "Small Classes", ylab = "Residuals")
abline(h=0, col = "red")
identify(college$Pct.20, fit$residuals, n=2, labels=college$School)
````
