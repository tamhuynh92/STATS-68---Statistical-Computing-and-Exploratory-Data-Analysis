### 2.4 Time Series Data

```r
nhtemp
```

```r
plot(nhtemp)
```

```r
plot(nhtemp, ylab = "Mean annual temperatures")
abline(h = mean(nhtemp))
lines(lowess(nhtemp))
```

For modeling, one often wishes to transform a time series so that the mean is stable over time. If X1, X2, ... is the time series, then we can ontain the time series of first differences X2 - X1, X3 - X2, ... using __diff function__

```r
diff(nhtemp)
```

```r
d = diff(nhtemp)
plot(d, ylab = "First differences of mean annual temperature")
abline(h = 0, lty = 3)
line(lowess(d))
```
