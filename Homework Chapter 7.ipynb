__7.5 (Hubble’s Law).__ <br> In 1929 Edwin Hubble investigated the relationship
between distance and velocity of celestial objects. Knowledge of this relationship might give clues as to how the universe was formed and what may
happen in the future. Hubble’s Law is is <br><br>
$$ Recession Velocity = H_{0} × Distance $$ <br>
where H0 is Hubble’s constant. This model is a straight line through the origin
with slope H0. Data that Hubble used to estimate the constant H0 are given
on the DASL web at http://lib.stat.cmu.edu/DASL/Datafiles/Hubble.
html. Use the data to estimate Hubble’s constant by simple linear regression.

```r
hubble = read.table("hubble.txt", header = TRUE)
hubble.lm = lm(recvelocity ~ 0 + distance, data = hubble)
hubble.lm
plot(recvelocity ~ distance, data = hubble, ylim = c(0,1000), xlim = c(0,2))
abline(0, 423.9, col = "red")
```
