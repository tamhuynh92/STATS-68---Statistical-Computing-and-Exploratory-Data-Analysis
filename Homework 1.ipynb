*** 1.4 (Binomial probabilities). *** Let X be the number of ones obtained in 12 rolls of a fair die. Then X has a Binomial(n = 12, p = 1/3) distribution. Compute a table of binomial probabilities for x = 0,1,...,12 by two methods: 

a. Use the probability density formula and vectorized arithmetic in R. Use 0:12 for the sequence of x values and
the ***choose function*** to compute the binomial coefficients.


b. Use the dbinom function provided in R and compare your results using both methods

```r
#a 

k <- c(0:12)
n <- 12
p <- 1/3

Prob <- ((factorial(n))/(factorial(n-k)*factorial(k)))*(p^k)*(1-p)^(n-k)

mytable = matrix(0, 1, length(k), dimnames = list(c("Prob"), k))
mytable["Prob",] = round(Prob*100, digit = 5)
mytable
```

```r
#b
mytable = matrix(0, 1, length(k), dimnames = list(c("Prob"), k))
mytable["Prob",] = round(dbinom(k, size = 12, prob = 1/3)*100, digit = 5)
mytable
```

***1.5 (Binomial CDF).*** Let X be the number of ones obtained in 12 rolls of a fair die. Then X has a Binomial(n = 12, p = 1/3) distribution. Compute a table of cumulative binomial probabilities (the CDF) for x = 0,1,...,12 by two methods: (1) using cumsum and the result of Exercise 1.4, and (2) using the pbinom function. What is P(X > 7)?

```r
k <- c(0:12)
n <- 12
p <- 1/3

Prob <- ((factorial(n))/(factorial(n-k)*factorial(k)))*(p^k)*(1-p)^(n-k)
sum = rep(NA, 12)
for(i in 1:length(Prob))  
{
    sum[i] = sum(Prob[1:i])
}

mytable = matrix(0,1, length(k), dimnames = list(c("Cum. Prob %"), k))
mytable["Cum. Prob %",] = round(sum*100, digit = 4)
mytable
```

```r
#b
mytable = matrix(0, 1, length(k), dimnames = list(c("Cum. Prob %"), k))
mytable["Cum. Prob %",] = round(pbinom(k, size = 12, prob = 1/3)*100, digit = 5)
mytable
```

```r
paste("P>7:", round(1 - pbinom(7, size = 12, prob = 1/3), digit = 5))
paste("P>7:", round(pbinom(7, size = 12, prob = 1/3, FALSE), digit = 5))
```

*** 1.7 (Simulatedhorsekicksdata).*** The rpois function generates random observations from a Poisson distribution. In Example 1.3, we compared the deaths due to horsekicks to a Poisson distribution with mean  = 0.61, and in Example 1.4 we simulated random Poisson( = 0.61) data. Use the rpois function to simulate very large (n = 1000 and n = 10000) Poisson( = 0.61) random samples. Find the frequency distribution, mean and variance for the sample. Compare the theoretical Poisson density with the sample proportions (see Example 1.4)

```r
y = rpois(1000, lambda = 0.61)
table(y) #Frequency
table(y)/1000 #Sample Proportion
cat("Mean =", mean(y))
cat("\nVar =", var(y))

Theoretical = dpois(0:4, lambda=.61)
Sample = table(y)/1000
cbind(Theoretical, Sample)
```

```r
y = rpois(10000, lambda = 0.61)
table(y) #Frequency
table(y)/10000 #Sample Proportion
cat("Mean =", mean(y))
cat("\nVar =", var(y))

Theoretical = dpois(0:6, lambda=.61)
Sample = table(y)/10000
cbind(round(Theoretical,4), Sample)
```
